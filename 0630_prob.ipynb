>### 직관
>#### <br>`0` 요지는 결측치 처리인듯 하다
>타이타닉 사건 때 노인과 어린이들이 먼저 구명보트에 탔다는 이야기를 들었다. 이러면 일단 나이는 생존 여부를 가르는 중요한 변수인데, 이 나이 데이터가 대략 20% 정도 날라갔다. 따라서 이 나이 변수를 어떻게든 채워넣어야 하지 않을까? 만약 나이 변수가 다른 변수와 연관성이 있다면 다른 변수들로 회귀분석을 하는 식으로 채우는 것도 가능하다.
>
>#### <br>`1` 결국 우리가 할 건 분류
>결국 요지는 새로운 데이터를 줬을 때 이게 생존자 그룹인지 사망자 그룹인지 구분하는 것이다. 따라서 결국은 DA나 SVM를 써야 할 거 같다.
>
>#### <br>`2` 변수 축소를 해야 하는데, 주성분을 찾을지 아니면 새로운 성분을 정의할지 결정해야 한다
>데이터 테이블에 사망 변수를 제외해도 변수가 11가지이다. 너무 많다. 일단 고윳값을 보고 무슨 변수가 중요한지 보자.
>
>#### <br>`3` 심슨의 역설을 피하기 위해서 인과관계도를 그려야 할 거 같다.

>### 따라서
>#### `0` 일단 그래프부터 그려보자 + 인과관계도를 생각해보자<br>
>#### `1` 고윳값보고 어느 변수가 중요한지 보자
>#### <br>`2` Age가 중요한 변수이면 다중회귀분석 식을 세우자
>#### <br>`3` 결측치를 보완한 데이터로 DA를 해보자
